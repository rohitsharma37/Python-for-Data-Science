````python
# Example 1
import math
class circle:
  def __init__(self,radius):
    self.radius=radius
  def setRadius(self,new_radius):
    self.radius=new_radius
  def area(self):
    ar=math.pi*((self.radius)**2)
    return ar
  def circumference(self):
    cm=2*math.pi*self.radius
    return cm
````

````python
# Example 2 - Rectangle
class rectangle:
  def __init__(self,length,breadth):
    self.length=length
    self.breadth=breadth
  def set_new(self,new_length,new_breadth):
    self.length=new_length
    self.breadth=new_breadth
  def area(self):
    ar=self.length*self.breadth
    return ar
  def perimeter(self):
    pr=2*(self.length+self.breadth)
    return pr
  def diagonal(self):
    dg=((self.length**2) + (self.breadth**2))**(1/2)
    return dg
````

````python
rect=rectangle(4,2)
````

# Inheritance

````python
# Parent Class
class aadhar():
  def __init__(self,name,aadhar_number):
    self.name=name
    self.aadhar_number=aadhar_number
  def display(self):
    print(self.name)
    print(self.aadhar_number)
  def info(self):
    print(f"Name : {self.name}")
    print(f"Aadhar Number : {self.aadhar_number}")

# Child Class
class employee(aadhar):
  def __init__(self,name,aadhar_number,salary,position):
    self.salary=salary
    self.position=position
  # Initiating the attributes of parent class
    aadhar.__init__(self,name,aadhar_number)
  def details(self):
    return print(f"Name : {self.name}, \nID/Aadhar : {self.aadhar_number}, \nRole : {self.position}, \nCompensation : {self.salary} ")
````

````python
a = employee( 'Rahul', 886012, 200000, "Intern")
a.details() # We call a method from child class.
````

````python
b = employee('Akshay', 886011, 200000, "Intern")
b.info()   # We can call a method from the parent class.
````

# **else**

### In python, you can also use the else clause on the try-except block which must be present after all the except clauses. The code enters the else block only if the try clause does not raise an error. We can use the result from the try clause to execute the code in the else block.

````python
def div(x,y):
  try:
    z=x/y
    print("No error")
  except ZeroDivisionError:
    print("Cannot divide by zero")
  else:
    return z
````

````python
div(5,9)
````

````python
div(3,0)
````

# **finally**

## will get executed no matter what

````python
def div(x,y):
  try:
    z=x/y
    print("No error")
  except ZeroDivisionError:
    print("Cannot divide by zero")
  else:
    return z
  finally:
    print("will get executed no matter what")
````

# Real World Use Case

### Uber's simplified pricing model

When you request an Uber, you enter your pick-up location and the destination. Based on the distance, peak hours, willingness to pay and many other factors, Uber uses a machine learning algorithm to compute what prices will be shown to you.

Let's consider a simplistic version of the Uber Pricing Model where you compute the price based on the distance between the pick-up and the drop location and the time of the booking.

User Inputs:

* Pick-up location (pick_up_latitude, pick_up_longitude)
* Drop location (drop_latitude, drop_longitude)
* Time of booking

Output:

* Final Price

# Development Code

````python
# Calculating distance between pick-up and drop locations

def get_distance(location1,location2):
  
  import geopy.distance
  
  distance=geopy.distance.distance(location1,location2).km
  
  return distance
````

````python
# Now getting price per km 

def get_price(hour):
  
  if hour>8 and hour<11:
   
    price = 20
  
  elif hour>18 and hour<21:
    
    price=15
  
  else:
    
    price=10
 
  return price
````

````python
# Now calculating the final price using the above two functions

def get_final_price(pickup_location,drop_location,booking_hour):
  
  total_distance=get_distance(pickup_location,drop_location)
  
  price_per_km = get_price(booking_hour)
  
  final_price = round(total_distance*price_per_km,2)
  
  return final_price
````

````python
# Calculate the distance between the pick-up location and the drop location.

# Every method we wirte, we write it in the class. So all the related functions stay in one place itself which makes it easy to comprehend.

class Maps:
  
  import geopy.distance
  
  import math
  
  def __init__(self):
    
    pass
  
  def get_distance(self,location1,location2):
   
    try:
     
      distance=geopy.distance.distance(location1,location2).km
    
    except:
      
      distance = math.sqrt((location1[0]-location2[0])^2 + (location1[1]-location2[1])^2)
    
   
    return distance
````

````python
class surgepricing:
  
  def __init__(self):
    pass
  
  def get_price(self,hour):
    
    try:
      
      if hour>8 and hour<11:
        price=20
      
      elif hour>18 and hour<21:
        price=15 
      
      else:
        price=10
    
    except:
      price=10
    
    return price
````

````python
def get_final_price(pickup_location,drop_location,booking_hour):
  
  maps=Maps() # Creating instance
  
  surge=surgepricing()

  
  total_distance=maps.get_distance(pickup_location,drop_location)
  
  actual_price= surge.get_price(booking_hour)

  
  final_price= round(total_distance * actual_price,2)
  
  return final_price
````
